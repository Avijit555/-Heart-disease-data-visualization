```python
import pandas as pd
import numpy as np
import seaborn as sns
df=pd.read_csv("heart.csv")
df.head()
```

## distribution plot for numeric column

```python
sns.displot(df['Age'],kde=True)
```

```python
sns.distplot(df['Age'],kde=True,hist=True,rug=True)
```

```python
sns.distplot(df['Age'],bins=30)
```

```python
sns.displot(df['Age'],kde=True,color ='blue')
```

```python
sns.displot(df['RestingBP'],kde=True,color ='blue')
```

```python
sns.distplot(df['RestingBP'],kde=True,color ='blue')
```

```python
sns.displot(df['Cholesterol'],kde=True,color ='blue')
```

```python
sns.displot(df['FastingBS'],kde=True,color ='blue')
```

## pie charts for categorical

```python
df.groupby('Sex').size().plot(kind='pie',autopct="%0.1f")
```

```python
df.groupby('ChestPainType').size().plot(kind='pie',autopct="%0.1f")
```

## violenplot

```python
sns.violinplot(x=df['Age'])
```

```python
sns.violinplot(y=df['Age'])
```

```python
sns.violinplot(x='Sex', y='HeartDisease', hue='Sex', data=df)
```

```python
sns.violinplot(y=df['Sex'], x=df['HeartDisease'])
```

## correlation heatmap 

```python
df.select_dtypes(include=['number']).corr()
```

```python
sns.heatmap(df.select_dtypes(include=['number']).corr())
```

## jointplot

```python
sns.jointplot(x='Age', y='MaxHR', data=df,kind='hex')
```

```python
sns.jointplot(x='Age', y='MaxHR', data=df,kind='reg')
```

```python
sns.jointplot(x='Age', y='Age', data=df)
```

```python
sns.jointplot(x='HeartDisease', y='Oldpeak', data=df)
```
